```python
import pandas as pd
import numpy as np

customers = pd.read_csv('../data/raw.csv')
customers.head()
```

```python
customers.info()
```

```python
"Percentage of customers churned: %{:.2f}".format(len(customers[customers['Churn']=='Yes'])/len(customers)*100)
```

```python
# Change object column to float
customers['TotalCharges'] = pd.to_numeric(customers.TotalCharges, errors='coerce')

# Check for NaN resulting from error coercion
customers[customers.TotalCharges.isnull()]
```

```python
# All NaN TotalCharges have 0 tenure, but have not churned; must be brand-new customers who haven't paid 1st bill
customers = customers.fillna(value={'TotalCharges':0})
```

```python
# Check float columns
customers[['tenure','MonthlyCharges','TotalCharges']].describe()
```

```python
# Looks like we have a lot of Yes/No and other categorical columns. Check unique values:
uniques = {col:customers[col].unique() if len(customers[col].unique())<=10 else '>10 unique values' \
           for col in customers.columns}
uniques_df = pd.DataFrame.from_dict(uniques, orient='index', columns=['unique values'])
uniques_df
```

```python
# PaymentMethod values are cut off in the above view. Closer look:
customers.PaymentMethod.unique()
```

```python
# Make SeniorCitizen consistent with other categorical features
customers['SeniorCitizen'] = pd.Series(np.where(customers['SeniorCitizen'].values == 1, 'Yes', 'No'))
```

```python
customers[customers.duplicated()]
```

```python
customers.customerID.is_unique
```

I have no duplicates, I have a unique identifier column, I have no missing values, and the data types all appear to be appropriate. Later, I will have to further process the categorical columns (e.g., creating dummies) but from a data quality standpoint, everything seems to be in order here.
